```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50]).reshape(-1, 1) 
y = np.array([25, 30, 40, 45, 50, 60, 65, 70, 80])               

model = LinearRegression()
model.fit(x, y)

m = model.coef_[0]
b = model.intercept_
print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Equation: y = {m:.2f}x + {b:.2f}")

x_predict = 60
y_predict = m * x_predict + b
print(f"Predicted sales growth for 60k investment: {y_predict:.2f}k")

desired_growth = 100
x_needed = (desired_growth - b) / m
print(f"Investment needed for 100k growth: {x_needed:.2f}k")

#  
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, model.predict(x), color='red', label='Regression line')
plt.scatter([x_predict], [y_predict], color='green', s=100, label='Prediction for 60k')
plt.scatter([x_needed], [desired_growth], color='purple', s=100, label='Investment for 100k')

plt.title('Linear Regression - Advertising Investment vs Sales Growth')
plt.xlabel('Advertising Investment (1000 ILS)')
plt.ylabel('Sales Growth (1000 ILS)')
plt.grid(True)
plt.legend()
plt.text(9, 82, f'y = {m:.2f}x + {b:.2f}', fontsize=12)
plt.show()
```
